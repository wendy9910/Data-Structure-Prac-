```python
def factorial(n):
    if n==1:
        return 1
    else:
        return (n*factorial(n-1))

N=eval(input("輸入資料個數:"))
com = factorial(N)
print("資料個數:%d ,數列組合數 = %d" %(N,com))
```

```python
#建立鏈結串
class Node():  
    def __init__(self,data=None):
        self.data = data           #資料
        self.next = None           #指標

n1 = Node(5)                       #節點1
n2 = Node(15)                      #節點2
n3 = Node(25)                      #節點3
n1.next = n2                       #節點1指向節點2
n2.next = n3                       #節點2指向節點3
ptr = n1                           #建立指標節點

while ptr:
    print(ptr.data)                #列印指標節點
    ptr=ptr.next                   #移動指標到下一個節點
```
